# ANÃLISIS

```python
properati_gba =GBA.loc[:,['price_usd_per_m2','property_type','FECHA']]
properati_caba = capitalFederal.loc[:,['price_usd_per_m2','property_type','FECHA']]
properati_gba.dropna(inplace=True)
properati_caba.dropna(inplace=True)
properati_caba.count()
```

```python
#VariaciÃ³n precio de distintas propiedades a lo largo de los aÃ±os para el Gran Buenos Aires
%matplotlib inline

fig = plt.figure(figsize=(10,4));

ax = fig.add_axes([0,0,1,1]);
properati_gba.loc[properati_gba['property_type'] == "house"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='r', label="house");
properati_gba.loc[properati_gba['property_type'] == "apartment"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='b', label="apartment");
properati_gba.loc[properati_gba['property_type'] == "PH"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='g', label="PH");
properati_gba.loc[properati_gba['property_type'] == "store"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='y', label="store");
plt.title('Precios de diferentes tipos de propiedades a lo largo del tiempo', fontsize=23);
plt.xlabel('Fecha', fontsize=16);
plt.ylabel('Precio por metro cuadrado en dÃ³lares', fontsize=16);
ax.legend(fontsize=13);
```

```python
#VariaciÃ³n precio de distintas propiedades a lo largo de los aÃ±os para Capital Federal
%matplotlib inline

fig = plt.figure(figsize=(10,4));

ax = fig.add_axes([0,0,1,1]);
properati_caba.loc[properati_caba['property_type'] == "house"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='r', label="house");
properati_caba.loc[properati_caba['property_type'] == "apartment"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='b', label="apartment");
properati_caba.loc[properati_caba['property_type'] == "PH"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='g', label="PH");
properati_caba.loc[properati_caba['property_type'] == "store"].groupby('FECHA').mean().loc[:,'price_usd_per_m2'].plot.line(linewidth=2,figsize=(12,8), c='y', label="store");
plt.title('Precios de diferentes tipos de propiedades a lo largo del tiempo', fontsize=23);
plt.xlabel('Fecha', fontsize=16);
plt.ylabel('Precio por metro cuadrado en dÃ³lares', fontsize=16);
ax.legend(fontsize=13);
```

Se puede notar con estos dos grÃ¡ficos que en el Gran Buenos Aires son los almacenes los que maÃ¡s han variado su precio en el Ãºltimo tiempo, seguidos por los departamentos, mientras que las casas y los PHs se han mantenido dentro de todo mÃ¡s constantes. Luego, en cuanto a la Capital Federal, a diferencia del Gran Buenos Aires lo que ha mostrado un gran golpe en cuanto al aumento de su precio son los PHs, en el mes de Mayo de 2016, seguido luego por los departamentos que tambiÃ©n muestran un aumento notable en su precio en algunos meses, tanto en 2016 como 2017. Por otro lado las casas y los almacenes mantuvieron mejor su precio a lo largo de los aÃ±os. Cabe destacar que tanto en Capital Federal como en el Gran Buenos Aires, los que siempre mantuvieron un precio no tan variable fueron las casas.
